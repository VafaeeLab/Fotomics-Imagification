```python
import h5py
import numpy as np
import pandas as pd
```

```python
h5f = h5py.File('/content/drive/MyDrive/combined/sorted_h5_data/Diaphragm-counts_x_train.h5','r')
x_train = h5f['dataset_1'][:]
h5f.close()


h5f = h5py.File('/content/drive/MyDrive/combined/sorted_h5_data/Diaphragm-counts_x_test.h5','r')
x_test = h5f['dataset_1'][:]
h5f.close()

h5f = h5py.File('/content/drive/MyDrive/combined/sorted_h5_data/Diaphragm-counts_y_train.h5','r')
y_train = h5f['dataset_1'][:]
h5f.close()

h5f = h5py.File('/content/drive/MyDrive/combined/sorted_h5_data/Diaphragm-counts_y_test.h5','r')
y_test = h5f['dataset_1'][:]
h5f.close()
```

```python
x_train = np.nan_to_num(x_train)
x_test = np.nan_to_num(x_test)
sum_train = np.sum(x_train,axis=1)
sum_train = sum_train.reshape(sum_train.shape[0],-1)
sum_test = np.sum(x_test,axis=1)
sum_test = sum_test.reshape(sum_test.shape[0],-1)


x_train = np.divide(x_train, sum_train)
x_test =  np.divide(x_test, sum_test)
print(x_train.shape)
```

```python
import gc

del sum_train
del sum_test

gc.collect()
```

```python
unique_classes = np.unique(y_train)
```

```python
label = unique_classes.tolist()


n_lst=[]
for i in range(len(y_train)):
  for j in range(len(label)):
    if y_train[i] == label[j]:

      n_lst.append(j)

y_new_train = np.array(n_lst,dtype='float64')

n_lst=[]
for i in range(len(y_test)):
  for j in range(len(label)):
    if y_test[i] == label[j]:
      n_lst.append(j)

y_new_test = np.array(n_lst,dtype='float64')
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(max_depth=2, random_state=0)
clf.fit(x_train,y_new_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

out_acc = accuracy_score(y_new_test, y_pred)
print("Raw data Accuracy: {}".format(round(out_acc,4)))

out_f1 = f1_score(y_new_test, y_pred,average ='micro')
print("Raw data F1 Score(micro): {}".format(round(out_f1,4)))

out_f1 = f1_score(y_new_test, y_pred,average ='macro')
print("Raw data F1 Score(macro): {}".format(round(out_f1,4)))

out_f1 = f1_score(y_new_test, y_pred,average ='weighted')
print("Raw data F1 Score(weighted): {}".format(round(out_f1,4)))
```
